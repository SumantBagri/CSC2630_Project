## Solution Cost vs Execution Time

```python
fig, axs = plt.subplots(nrows=8, ncols=5, figsize=(15,10))

for idx in range(2,40):
    mapi_data = ddf[(ddf["Map Id"] == idx) & (ddf["Any Path Found"])]
    scatterdf = mapi_data[["Timestep", "Current Path Cost"]]
    linedf = mapi_data[["Timestep", "Current Path Cost", "Batch Size"]].groupby(["Batch Size"]).mean()
    axs[idx//5, idx%5].scatter(scatterdf["Timestep"], scatterdf["Current Path Cost"])
    axs[idx//5, idx%5].plot(linedf["Timestep"], linedf["Current Path Cost"])
fig.delaxes(axs[0, 0])
fig.delaxes(axs[0, 1])
fig.tight_layout()
plt.show()
```
