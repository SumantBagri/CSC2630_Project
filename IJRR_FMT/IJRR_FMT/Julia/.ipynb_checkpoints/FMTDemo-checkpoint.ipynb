```julia
include("BasicPlotting.jl")
include("MotionPlanning.jl")
include("ObstacleGenerators.jl")
```

```julia
obs = RandomNonIntersectingBoxes(2, .4)
init = .5*ones(2)
goal = BallGoal(ones(2), .1)

P = GeometricProblem(init, goal, obs, Array(Vector{Float64}, 0), UnitHypercube(2));
```

## FMT*

```julia
N = 1000
gc(); @time c, sol, V, A = FMTstar(P, N, 1.0, kNN=true, kdt=true)

PlotProblemSetup(P)
PlotTree(V, A, color="gray", alpha=0.5)
PlotSolution(V, sol, "blue")
c
```

## PRM*

```julia
N = 500
gc(); @time c, sol, V, F = PRMstar(P, N, 1.0, kNN=true, kdt=true)

PlotProblemSetup(P)
PlotGraph(V, F, color="gray", alpha=0.5)
PlotSolution(V, sol, "blue")
```

## RRT*

```julia
N = 1000
gc(); @time c, sol, V, A = RRTstar(P, N, .15, 1.0)

PlotProblemSetup(P)
PlotTree(V, A, color="gray", alpha=0.5)
PlotSolution(V, sol, "blue")
c
```
